```python
import cv2
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import struct
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten
from keras.datasets import mnist
from sklearn import datasets
from skimage.measure import regionprops
from skimage.measure import label
import tensorflow as tf
import math
```

```python
def load_data(): 
    with open("train-images.idx3-ubyte",'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.fromstring(f.read(), dtype=np.uint8).reshape(shape)
    
def load_label():
    with open("train-labels.idx1-ubyte", 'rb') as f:
        zero, data_type, dims = struct.unpack('>HBB', f.read(4))
        shape = tuple(struct.unpack('>I', f.read(4))[0] for d in range(dims))
        return np.fromstring(f.read(), dtype=np.uint8).reshape(shape)

numbers = load_data()
labels = load_label()
numbers = np.reshape(numbers, (60000, 28*28))


close_kernel = np.ones((5, 5), np.uint8)
raspon = range(0, len(numbers))
   
for i in raspon:
    digit = numbers[i].reshape(28, 28)
    upada = cv2.inRange(digit, 150, 255)
    closing = cv2.morphologyEx(upada, cv2.MORPH_CLOSE, close_kernel)
    labeled = label(closing)
    regions = regionprops(labeled)
    if(len(regions) > 1):
        max_width = 0
        max_height = 0
        for j in range(0, len(regions)):
            num = regions[j].bbox
            if(num[3] - num[1] > max_width and  num[2] - num[0] > max_height):
                max_height = num[3] - num[1]
                max_width = num[2] - num[0]
                bbox = num
    else:
        bbox = regions[0].bbox
    digit_pic = np.zeros((28, 28))
    x = 0
    for w in range(bbox[0], bbox[2]):
        y = 0
        for h in range(bbox[1], bbox[3]):
            digit_pic[x, y] = digit[w, h]
            y += 1
        x += 1
    numbers[i] = digit_pic.reshape(1, 28*28)

print("KNN Start")
knn = KNeighborsClassifier(n_neighbors=5, algorithm='auto').fit(numbers, labels)
print("KNN End")
```

```python
def blueLineDetection(frame):
    blueLight = np.array([110,50,50])
    blueDark = np.array([130,255,255])
    
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    blue = cv2.inRange(hsv, blueLight, blueDark)
    blueEdges = cv2.Canny(blue, 75, 150)

    blueLine = cv2.HoughLinesP(blueEdges, 1 ,np.pi/180, 50, maxLineGap=50)
    for x1,y1,x2,y2 in blueLine[0]:
        return x1, y1, x2, y2
```

```python
def greenLineDetection(frame):
    greenLight = np.array([45,100,50])
    greenDark = np.array([100,255,255])
    
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    green = cv2.inRange(hsv, greenLight, greenDark)
    greenEdges = cv2.Canny(green, 75, 150)

    greenLine = cv2.HoughLinesP(greenEdges, 1 ,np.pi/180, 50, maxLineGap=50)
    for x1,y1,x2,y2 in greenLine[0]:
        return x1, y1, x2, y2
```

```python
def find_numbers(entry_frame,pointStartB,pointEndB,pointStartG,pointEndG,knn,blueDigits, greenDigits,suma):
   
    xl1b = pointStartB[0]
    yl1b = pointStartB[1]
    xl2b = pointEndB[0]
    yl2b = pointEndB[1]
    
    xl1g = pointStartG[0]
    yl1g = pointStartG[1]
    xl2g = pointEndG[0]
    yl2g = pointEndG[1]
    
    startBlue = np.array([xl1b,yl1b])
    endBlue = np.array([xl2b,yl2b])
    
    startGreen = np.array([xl1g,yl1g])
    endGreen = np.array([xl2g,yl2g])
    
    gray_frame = cv2.cvtColor(entry_frame, cv2.COLOR_RGB2GRAY)
    ret,bin_frame = cv2.threshold(gray_frame, 127, 255, cv2.THRESH_BINARY)
    
    img, contours, hierarchy = cv2.findContours(bin_frame.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for con in contours:
        x,y,w,h = cv2.boundingRect(con) 
        number = np.zeros((28, 28))
        if w > 2 and w < 30 and h > 5 and h < 30: 
               
            center = []
            center.append(int(x + w / 2))
            center.append(int(y + h / 2)) 
            
    
            
            numberCenter = np.array([center[0],center[1]])
    
            
            if center[0]>startBlue[0] and center[0]<endBlue[0]:
                d = (center[0] - xl1b)*(yl2b-yl1b) - (center[1]-yl1b)*(xl2b-xl1b)
                if d > 0 :
                    for i in range(0,h):
                        for j in range(0, w):
                            number[i, j] = gray_frame[y+i+1,x+j+1]
                
                    distance=np.cross(startBlue-endBlue,numberCenter-endBlue)/np.linalg.norm(startBlue-endBlue)
                    
                    if(distance > 13 and distance < 16):
                        br = int(knn.predict(number.reshape(1,784)))
                        cv2.putText(entry_frame, str(int(br)), (x, y-5),cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 0), 1)
                       
                        foundNumbers = []
                        for i,el in enumerate(blueDigits):
                            x = el[1][0]
                            y = el[1][1]
                            distanca = math.sqrt(math.pow((x - center[0]), 2) + math.pow((y - center[1]), 2))
                            if distanca < 15 and el[0]==int(br):
                                foundNumbers.append(i)
                        if len(foundNumbers) == 0 :
                            print("Prvi put " + str(int(br)))
                            blueDigits.append((int(br), center))
                            suma  = suma + int(br)
                            #print("Suma " + str(suma))
                        elif len(foundNumbers)==1:
                           # print("Dugi put " + str(int(br)))
                            index = foundNumbers[0]
                            blueDigits[index][1][0] = center[0]
                            blueDigits[index][1][1] = center[1]
                        
            
    
            if center[0]>startGreen[0] and center[0]<endGreen[0]:
                d = (center[0] - xl1g)*(yl2g-yl1g) - (center[1] - yl1g)*(xl2g-xl1g)
                if d > 0:
                    for i in range(0,h):
                        for j in range(0, w):
                            number[i, j] = gray_frame[y+i+1,x+j+1]
                    
                    distance = np.cross(startGreen-endGreen,numberCenter-endGreen)/np.linalg.norm(startGreen-endGreen)
                    
                    if(distance > 13 and distance < 16):
                        br = int(knn.predict(number.reshape(1,784)))
                        cv2.putText(entry_frame, str(int(br)), (x, y-5),cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 0), 1)
                    
                        foundNumbers = []
                        for i,el in enumerate(greenDigits):
                            x = el[1][0]
                            y = el[1][1]
                            distanca = math.sqrt(math.pow((x - center[0]), 2) + math.pow((y - center[1]), 2))
                            if distanca < 15 and el[0]==int(br):
                                foundNumbers.append(i)
                        if len(foundNumbers) == 0 :
                            print("Prvi put zelena" + str(int(br)))
                            greenDigits.append((int(br), center))
                            suma  = suma - int(br)
                            #print("Suma " + str(suma))
                        elif len(foundNumbers)==1:
                            #print("Dugi put " + str(int(br)))
                            index = foundNumbers[0]
                            greenDigits[index][1][0] = center[0]
                            greenDigits[index][1][1] = center[1]
                        
           
            
    return entry_frame, suma
```

```python


cap = cv2.VideoCapture('video-0.avi')
ret, frame = cap.read()
x1B, y1B, x2B, y2B = blueLineDetection(frame)
x1G, y1G, x2G, y2G = greenLineDetection(frame)
pointStartB = []
pointStartB.append(x1B)
pointStartB.append(y1B)
pointEndB = []
pointEndB.append(x2B)
pointEndB.append(y2B)

blueDigits = []
greenDigits = []


pointStartG = []
pointStartG.append(x1G)
pointStartG.append(y1G)
pointEndG = []
pointEndG.append(x2G)
pointEndG.append(y2G)

suma = 0

while(cap.isOpened()):
    ret, frame = cap.read()
    if not ret : 
        break
    #cv2.line(frame,(x1B,y1B),(x2B,y2B),(0,255,255),3)
    #cv2.line(frame,(x1G,y1G),(x2G,y2G),(0,255,255),3)
    frame, suma = find_numbers(frame.copy(),pointStartB,pointEndB,pointStartG,pointEndG,knn, blueDigits, greenDigits,suma)
    #frame, suma = find_numbers(frame.copy(),pointStartB,pointEndB,knn, blueDigits,suma)

    
    cv2.putText(frame, "Sum: " + str(suma), (10, 20),cv2.FONT_HERSHEY_DUPLEX, 1, (255, 0, 0), 1)
    
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
        
print("Suma: " + str(suma))
cap.release()
cv2.destroyAllWindows()
```
